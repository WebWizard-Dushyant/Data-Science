```python
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df[['job_title_short', 'job_skills']].head(5)
```

```python
df_exploded = df.explode('job_skills')
df_exploded 
```

```python
df_exploded['job_skills'].value_counts().head(10).plot(kind='bar')  
```

```python
skills_count = df_exploded.groupby(['job_skills','job_title_short']).size()
```

```python
df_skills_count = skills_count.reset_index(name='skills_count')
```

```python
df_skills_count
```

```python
df_skills_count = df_skills_count.sort_values(by='skills_count', ascending=False)
```

```python
df_skills_count
```

```python
job_title = "Data Analyst"
top_skills = 10

df_skill_final = df_skills_count[df_skills_count['job_title_short'] == job_title].head(top_skills)
print(df_skill_final)
```

```python
job_title = "Data Analyst"
top_skills = 10
df_skill_final = df_skills_count[df_skills_count['job_title_short'] == job_title].head(top_skills)

df_skill_final.plot(x='job_skills', y='skills_count', kind='barh')
plt.gca().invert_yaxis()
```
