# Classes

````python
class LukesList:
    def __init__(self):
        """Initialize an empty list."""
        self._items = []

    def add(self, item):
        """Add an item to the end of the list."""
        self._items.append(item)

    def __getitem__(self, index):
        """Retrieve an item by index."""
        return self._items[index]

    def __setitem__(self, index, value):
        """Set an item at a specific index."""
        self._items[index] = value

    def __repr__(self):
        """Return a string representation of the list."""
        return str(self._items)

    def __len__(self):
        """Return the length of the list."""
        return len(self._items)
````

# Classes

````python
class BaseSalary:
  def __init__(self, base_salary, bonus_rate=0.1, symbol="$"):
    self.base_salary = base_salary
    self.bonus_rate = bonus_rate
    self.symbol = symbol
    # Calculate the numerical total salary directly
    self._numerical_total_salary = self.base_salary * (1 + self.bonus_rate)
    # Calculate the numerical bonus amount directly
    self._numerical_bonus_amount = self.total_salary - self.base_salary

  def __repr__(self):
    # Returning the formatted total salary for representation
    return self.calculate_salary()

  def calculate_salary(self):
    # Returns the formatted total salary
    return  f"{self.symbol}{self._numerical_total_salary:,.0f}"

  def calculate_bonus(self):
    # Returns the formatted bonus amount
    return  f"{self.symbol}{self._numerical_bonus_amount:,.0f}"
````
